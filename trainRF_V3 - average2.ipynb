```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

# DataFrame
results_df = pd.DataFrame(columns=['ThomasAM_2018a', 'ThomasAM_2018b',"YuJex","ZellerGex","FengQex","VogtmannEex","HanniganGDex"],index=['ThomasAM_2018a', 'ThomasAM_2018b',"YuJex","ZellerGex","FengQex","VogtmannEex","HanniganGDex"])
```

```python

# 7
variables = [ThomasAM_2018a_c, ThomasAM_2018b_c, YuJex_c, ZellerGex_c, FengQex_c, VogtmannEex_c, HanniganGDex_c]

# 
for i in variables:
    # 
    
    # 
    X = i.drop('label', axis=1)
    y = i['label']
    
    # 20
    auc_scores = []
    for _ in range(20):
        # 
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=_)
        
        # 
        model = RandomForestClassifier()
        
        # 
        model.fit(X_train, y_train)
        
        # 
        y_pred = model.predict_proba(X_test)[:, 1]  # 
        
        # AUC
        auc = roc_auc_score(y_test, y_pred)
        
        # AUC
        auc_scores.append(auc)
    
    # AUC
    avg_auc = sum(auc_scores) / len(auc_scores)
    
    variable_name = [name for name, value in globals().items() if value is i][0][:-2]
    print(variable_name,"same")  
    
    # DataFrame
    results_df.loc[variable_name,variable_name] = avg_auc
    variable_temp = [x for x in variables if x.all != i.all]
    
    model = RandomForestClassifier()
    # 
    model.fit(X, y)
    for s in variable_temp:
        # 
        X_test = s.drop('label', axis=1)
        y_test = s['label']
        
        y_pred = model.predict_proba(X_test)[:, 1]  # 
        
        # AUC
        auc = roc_auc_score(y_test, y_pred)
        print(auc)
        variable_name_train = [name for name, value in globals().items() if value is i][0][:-2]
        variable_name_test = [name for name, value in globals().items() if value is s][0][:-2]
        print(variable_name_train,"train")
        print(variable_name_test,"test")
        results_df.loc[variable_name_train,variable_name_test] = auc
```

```python
from sklearn.linear_model import Lasso
import pandas as pd

# Lasso
def lasso_regression(data, alpha=0.01):
    target = data.iloc[:, -1]
    features = data.iloc[:, :-1]
    lasso = Lasso(alpha=alpha)
    lasso.fit(features, target)
    selected_features = features.columns[lasso.coef_ != 0]
    return selected_features

# DataFrame
selected_dict = {}

# DataFrame
for var_name, df in zip(['ThomasAM_2018a_c', 'ThomasAM_2018b_c', 'YuJex_c', 
                         'ZellerGex_c', 'FengQex_c', 'VogtmannEex_c', 'HanniganGDex_c'], variables):
    selected_features = lasso_regression(df, alpha=0.01)
    selected_dict[var_name] = selected_features
    
# DataFrame
for var_name, features in selected_dict.items():
    print(var_name)
    
    print("#"*50)
    #print('Original features:', list(variables[i].columns[:-1]))
    print('Selected features:', features)
    print("*"*20,"End","*"*20)
    print()
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn.model_selection import KFold
import numpy as np
def roc_auc_score_FIXED(y_true,y_pred):
    if len(np.unique(y_true)) == 1: # bug in roc_ auc_ score
        return accuracy_score(y_true, np.rint(y_pred))
    return roc_auc_score(y_true, y_pred)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

# DataFrame
results_df_newV2 = pd.DataFrame(columns=['ThomasAM_2018a', 'ThomasAM_2018b',"YuJex","ZellerGex","FengQex","VogtmannEex","HanniganGDex"],index=['ThomasAM_2018a', 'ThomasAM_2018b',"YuJex","ZellerGex","FengQex","VogtmannEex","HanniganGDex","LODO"])
```

```python
# 7
variables = [ThomasAM_2018a_c, ThomasAM_2018b_c, YuJex_c, ZellerGex_c, FengQex_c, VogtmannEex_c, HanniganGDex_c]
LODO = {}
# 
for i in variables:
    # 
    variable_name = [name for name, value in globals().items() if value is i][0][:-2]
    LODO[variable_name]=[]
    # 
    X = i.drop('label', axis=1)
    y = i['label']
    
    # 20
    auc_scores = []
    for _ in range(20):
        # 
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=_)
        
        # 
        model = RandomForestClassifier()
        
        # 
        model.fit(X_train, y_train)
        
        # 
        y_pred = model.predict_proba(X_test)[:, 1]  # 
        
        # AUC
        auc = roc_auc_score(y_test, y_pred)
        
        # AUC
        auc_scores.append(auc)
    
    # AUC
    avg_auc = sum(auc_scores) / len(auc_scores)
    
    variable_name = [name for name, value in globals().items() if value is i][0][:-2]
    print(variable_name,"same")  
    
    # DataFrame
    results_df_newV2.loc[variable_name,variable_name] = avg_auc
    variable_temp = [x for x in variables if x.all != i.all]
###########################################
##########################################LODO
    kf = KFold(n_splits=20) #
    auc_scores = []
    for train_index, val_index in kf.split(X):
        # 
        X_train, X_val = X.iloc[train_index], X.iloc[val_index]
        y_train, y_val = y.iloc[train_index], y.iloc[val_index]
        
        LODO[variable_name].append(y_val.index.to_list())
        
        # 
        model = RandomForestClassifier()
        
        # 
        model.fit(X_train, y_train)
        
        # 
        y_pred = model.predict_proba(X_val)[:, 1]  # 
        
        # AUC
        auc = roc_auc_score_FIXED(y_val, y_pred) #roc_auc_score 01
        #
        
        # AUC
        auc_scores.append(auc)
    
    # AUC
    avg_auc = sum(auc_scores) / len(auc_scores)
    
    print(variable_name, "LODO average AUC:", avg_auc)
    results_df_newV2.loc["LODO",variable_name] = avg_auc
#############################################    LODO
    
    
    model = RandomForestClassifier()
    # 
    model.fit(X, y)
    for s in variable_temp:
        # 
        X_test = s.drop('label', axis=1)
        y_test = s['label']
        
        y_pred = model.predict_proba(X_test)[:, 1]  # 
        
        # AUC
        auc = roc_auc_score(y_test, y_pred)
        print(auc)
        variable_name_train = [name for name, value in globals().items() if value is i][0][:-2]
        variable_name_test = [name for name, value in globals().items() if value is s][0][:-2]
        print(variable_name_train,"train")
        print(variable_name_test,"test")
        results_df_newV2.loc[variable_name_train,variable_name_test] = auc
```
